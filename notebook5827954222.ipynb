```python
df = pd.read_csv('/kaggle/input/ethereum-frauddetection-dataset/transaction_dataset.csv')
```

```python
df.drop([df.columns[0], df.columns[1], df.columns[2], df.columns[49],df.columns[50]], axis=1, inplace=True)
```

```python
df = df.dropna()
df = df.reset_index(drop = True)
df
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['FLAG']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
strat_test_set['FLAG'].value_counts()
```

```python
df = strat_train_set.copy()
```

```python
X = df.drop(['FLAG'], axis=1)
y = df['FLAG']
```

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,y_pred)
```
